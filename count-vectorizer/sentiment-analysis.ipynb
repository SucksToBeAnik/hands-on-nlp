```python
import pandas as pd
# dataset link: https://media.geeksforgeeks.org/wp-content/uploads/20240514105101/IMDB-Dataset.csv
df = pd.read_csv("../datasets/IMDB_movie_review_dataset.csv")
df.head()
len(df)
```

```python
import numpy as np

df["sentiment"] = np.where(df["sentiment"] == "positive", 1, 0)
```

```python
review = df["review"]
sentiment = df["sentiment"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(review, sentiment, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_train_transformed = vectorizer.fit_transform(X_train)
X_test_transformed = vectorizer.transform(X_test)
```

```python
X_train_transformed.shape
# print(X_test_transformed)
```

```python
from sklearn.naive_bayes import MultinomialNB

multinomial_model = MultinomialNB()
multinomial_model.fit(X_train_transformed, Y_train)
```

```python
from sklearn.metrics import accuracy_score
# accuracy
print(f"Training Accuracy: ${accuracy_score(Y_train, multinomial_model.predict(X_train_transformed))}")
print(f"Testing Accuracy: ${accuracy_score(Y_test, multinomial_model.predict(X_test_transformed))}")
```

```python
review_input = input("Enter a review: ")
review_input = [review_input]
review_input = vectorizer.transform(review_input)
print("Positive" if multinomial_model.predict(review_input) == 1 else "Negative") 
```

```python
# try a different model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_transformed, Y_train)

print(f"Training Accuracy: ${accuracy_score(Y_train, model.predict(X_train_transformed))}")
print(f"Testing Accuracy: ${accuracy_score(Y_test, model.predict(X_test_transformed))}")
```

```python
review_input = input("Enter a review: ")
review_input = [review_input]
review_input = vectorizer.transform(review_input)
print("Positive" if model.predict(review_input) == 1 else "Negative") 
```

### Comment
Logistic Regression Model performs better. Accuracy of 0.89 whereas MultinomialNB gave 0.84

```python
# Now perform nlp preprocessing
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.tag import pos_tag
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
```

```python
nltk.download("punkt_tab",download_dir="f:\\Work\\hands-on-nlp\\venv\\nltk_data")
nltk.download("wordnet",download_dir="f:\\Work\\hands-on-nlp\\venv\\nltk_data")
nltk.download("omw-1.4",download_dir="f:\\Work\\hands-on-nlp\\venv\\nltk_data")
nltk.download("averaged_perceptron_tagger_eng",download_dir="f:\\Work\\hands-on-nlp\\venv\\nltk_data")
```

```python
lemmatizer = WordNetLemmatizer()

def get_pos_given_word(word:str) :
    tag = pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

def custom_tokenizer(text):
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(word.lower(), get_pos_given_word(word)) for word in tokens]
    return lemmatized_tokens

custom_stop_words = set(lemmatizer.lemmatize(word.lower(), get_pos_given_word(word)) for word in ENGLISH_STOP_WORDS)

lemmatized_vectorizer = CountVectorizer(tokenizer=custom_tokenizer, stop_words=list(custom_stop_words))
```

```python
sentence = input("Enter a sentence: ")
test_fitted_model = lemmatized_vectorizer.fit([sentence])
print(ENGLISH_STOP_WORDS)
print(test_fitted_model.get_feature_names_out())
print(word_tokenize(sentence))
```

```python
X_train_transformed = lemmatized_vectorizer.fit_transform(X_train)
X_test_transformed = lemmatized_vectorizer.transform(X_test)
lemmatized_model = MultinomialNB()
lemmatized_model.fit(X_train_transformed, Y_train)
```

```python

print(f"Training Accuracy: ${accuracy_score(Y_train, lemmatized_model.predict(X_train_transformed))}")
print(f"Testing Accuracy: ${accuracy_score(Y_test, lemmatized_model.predict(X_test_transformed))}")
```

```python
review_input = input("Enter a review: ")
review_input = [review_input]
review_input = lemmatized_vectorizer.transform(review_input)
print("Positive" if lemmatized_model.predict(review_input) == 1 else "Negative") 
```
